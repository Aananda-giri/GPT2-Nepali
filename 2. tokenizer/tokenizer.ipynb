```python
#
from tokenizers import Tokenizer

# Load the pre-trained tokenizer from a file
tokenizer = Tokenizer.from_file("tokenizer_30k.json")

# Get the vocabulary and frequencies
vocab = tokenizer.get_vocab()  # Returns a dictionary {token: id}
vocab_size = len(vocab)
vocab_size
```

```python
tokenizer.model.unk_token = '<|unk|>'
enc=tokenizer.encode('<|endoftext|>')
print(enc.ids, enc.tokens)
```

```python
 sorted(vocab.items(), key=lambda item: item[1], reverse=True)
```
